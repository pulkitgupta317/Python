Import pandas library

```python
import pandas as pd
```

Loading the custom dataset from the local directory. I have used the Tips dataset for illustrating the data visualization. You can download the dataset from https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv.

```python
# loading the dataset from the csv file stored in the same directory as the Jupyter notebook file.
df = pd.read_csv('tips.csv')
```

```python
# printing the top 5 rows of the dataset
df.head()
```

# Scatter plot

Scatter plot is used to depict the distribution and relationship of the data based on two features/variables. Normally, we use plotting in scatter plot when we want know the trend of the data distribution between two numerical values. <a href='https://en.wikipedia.org/wiki/Scatter_plot#:~:text=A%20scatter%20plot%20(also%20called,for%20a%20set%20of%20data.'>more information</a>

```python
# x and y represent the feature/column on x-axis and y-axis respectively.
# data attribute is used to pass the dataframe from which the seaborn will create the plot.
sns.relplot(x="total_bill", y="tip", data=df)
```

We can add another feature as well to depict the distribution based on the three variables. For example:

```python
sns.relplot(x="total_bill", y="tip", hue='sex', data=df)
```

If we want to plot multiple charts based on a categorical feature then we can do it by passing the feature name in the 'col' attribute

```python
sns.relplot(x="total_bill", y="tip", hue='sex', col='day', data=df)
```

# Line Plot

Line plot is used when we want to understand the changes in one feature as a function of another feature. <a href='https://en.wikipedia.org/wiki/Line_chart'>more information</a>

```python
# we pass kind='line' when we want to plot a line chart, default value of kind is 'scatter'
sns.relplot(x="size", y="total_bill", kind="line", data=df)
```

```python
df.groupby('size')['total_bill'].mean()
```

Seaborn aggregate the multiple measurements at each x value by plotting the mean value.
For example, the mean value of total_bill feature by different values of size feature can be observed above.

We can add another feature as well to depict the distribution based on the three variables. For example:

```python
sns.relplot(x="size", y="total_bill", kind="line", hue='smoker', data=df)
```

If we want to plot multiple charts based on a categorical feature then we can do it by passing the feature name in the 'col' attribute

```python
sns.relplot(x="size", y="total_bill", kind="line", hue='smoker', col='day', data=df)
```

# Histogram Plot

Histogram charts convey data through the length or height of a bar, which makes the depiction of the distribution more accurate. <a href='https://en.wikipedia.org/wiki/Histogram'>more information</a>

### Univariate Histogram

```python
# Categorical feature
sns.displot(x="day", data=df)
```

```python
# Continuous feature
sns.displot(x="total_bill", data=df)
```

we can pass the number of bins while plotting a histogram where the x-axis values will be equally distributed in the number of bins.

```python
sns.displot(x="total_bill", bins=30, data=df)
```

### Multivariate Histogram

We can pass the feature name in the 'hue', it will draw a separate histogram for each of feature's unique values and distinguish them by the color.

```python
sns.displot(x="day", hue="smoker", data=df)
```

Instead of layering each bar, they can be stacked. This can be done by passing multiple='stack'

```python
sns.displot(x="day", hue="smoker", multiple='stack', data=df)
```

You may be thinking why the count in this chart are more for each value as compared to the previous chart. For that, please compare both charts.

If we want to plot multiple charts based on a categorical feature, we can do it by passing the feature name in the 'col' attribute

```python
sns.displot(x="day", hue="smoker", multiple='stack', col='sex', data=df)
```

# Box Plot

Box plots provide a quick visual summary of the distribution of values in a dataset. It shows the median, upper and lower quartiles, minimum and maximum values, and any outliers in the data. Outliers can reveal mistakes or unusual occurrences in data. <a href='https://en.wikipedia.org/wiki/Box_plot'>more information</a>

```python
sns.catplot(x="day", y="total_bill", kind="box", data=df)
```

We can further divide the x-axis distribution based on third variable as well using the 'hue' attribute of the catplot.

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=df)
```

If we want to plot multiple charts based on a categorical feature, we can do it by passing the feature name in the 'col' attribute

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", col='sex', data=df)
```
